Isolation Forest
( A negative score value and a -1 for the value of anomaly columns indicate the presence of anomaly. A value of 1 for the anomaly represents the normal data.)

```python
model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1),max_features=1.0)
model.fit(df_austria.iloc[:,0:2])
```

```python
df_austria['scores']=model.decision_function(df_austria.iloc[:,0:2])
df_austria['anomaly']=model.predict(df_austria.iloc[:,0:2])
df_austria.head(755)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a series of data of in range of 1-50.
x = df_austria.iloc[:,0:2]

#Creating a Function.
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)

#Apply function to the data.
pdf = normal_dist(x,mean,sd)

#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```
